```python
!pip3 install beautifulsoup4
import pandas as pd
import time
from bs4 import BeautifulSoup
from urllib import request
import re
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import TimeoutException
```

```python
url = 'https://snkhan.co.uk/stuff/iTunes.php?chart=US'
request_text = request.urlopen(url).read()

soup = BeautifulSoup(request_text, 'html.parser')
```

```python
def recuperer_infos_genre(soup):  
    all_products=[]
    index=1
    for d in soup.findAll('tr', attrs={'valign':'top'}):
        one_product=[]
        if (len(all_products)<100):
            classement= str(index)
            index+=1
            titre = d.find('td', attrs={'width':'95%'})
            for link in d.find_all("a", href=lambda href: href and href.startswith("https://music.apple.com/us/artist/")):
                artiste=link.text
            
            for link in d.find_all("a", href=lambda href: href and href.startswith("https://music.apple.com/us/genre/")):
                genre=link.text
            
            #.title() sert a mettre une majuscule au debut des noms
            one_product.append(classement)
            one_product.append(titre.a.text.title())
            one_product.append(artiste.title())
            one_product.append(genre.title())
            one_product.append("monde")
        
            all_products.append(one_product)
            
            
    return all_products
recuperer_infos_genre(soup)
```

```python
data_genre_monde=pd.DataFrame(recuperer_infos_genre(soup))
data_genre_monde.columns =['Classement', 'Titre','Artiste', 'Genre','Region']
data_genre_monde
```

```python
data_genre_monde.to_csv('genre_monde.csv', index=False)
```
